```python
# Cross-Section Updating Code 
# Written by Alec Granville-Willett
#
# This code imports new cross-section data into ARGO
# It interpolates the data to the correct bin size, splits it into individual photoreactions and imports 
# it into a format readable by ARGO
# 
# This is a working example for HCHO
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# This code imports the desired data set into a pandas dataframe.
ds = pd.read_csv('/home/ag834/Data/Mainz Sigma/Smith 300-340nm 294K', delimiter='\t') 
SmithSigma = abs(ds['sigma'])
```

```python
# As the ARGO bins for the cross-sections are 1 angstrom in size, the data must be redistributed into bins of this size.
# This code averages every x values into new values of the corresct bin size.

dsa = ds.iloc[::100, :]      # This line needs the number of values to be averaged over       
dsa['lambda'] = (dsa['lambda'] + 0.004)       # This line corrects the offset
for i in range(0, 383):
    dsa[i] = np.mean(ds['sigma'][i*100 : (i+1)*100])       # This line needs the same number
    
dsab = dsa.iloc[4]
RSmithSigma2 = dsab[2:386]        # This line needs to be set to 2:total number
```

```python
# This code puts the new data set into a dataframe.

dx = pd.DataFrame(np.zeros((383, 2)), columns =  ["lambda/A", "sigma/cm2"])      # This line needs the size of the new data
dx['lambda/A'] = np.arange(3000,3383).reshape((383, 1))                # This line needs the corresponding lambda values
dx['sigma/cm2'] = RSmithSigma2
dx.head()
```

```python
# As ARGO requires that each photreaction be seperate, the total cross-section data above needs to be split using the quantum yields.
# This code imports the quantum yield data, cuts out the appropriate wavelength range and interpolates the data to get 1 angstrom bins.

d28 = pd.read_csv('/home/ag834/Data/Mainz Sigma/H + HCO Phi', delimiter='        ')
d29 = pd.read_csv('/home/ag834/Data/Mainz Sigma/H2 + CO Phi', delimiter='        ')
dt = pd.read_csv('/home/ag834/Data/Mainz Sigma/Total Phi', delimiter='        ')
d30 = dt.copy()
d30['phi'] = dt['phi'] - d28['phi'] - d29['phi']                             
d28 = d28[50:90]      
d28 = d28.reset_index(drop=True)
d29 = d29[50:90]                          # These lines need the correct lambda values
d29 = d29.reset_index(drop=True)
d30 = d30[50:90]
d30 = d30.reset_index(drop=True)

zarry = pd.DataFrame(np.zeros((9, 2)), columns =  ["lambda", "phi"])         # This line needs a matrix of zeros of the corresct size
zarry[:] = np.nan                                                            # for the interpolation into 1 Angstrom bins

d28b = d28[:0]
for i in range(0, 40):                        # This line needs the size of the above dataframes
    d28b = d28b.append(d28.ix[i])
    d28b = d28b.append(zarry)
d28b = d28b.apply(pd.Series.interpolate)
d28b = d28b.reset_index(drop=True)
d28b = d28b[:383]                             # This line needs the correct size of the data

d29b = d29[:0]
for i in range(0, 40):
    d29b = d29b.append(d29.ix[i])
    d29b = d29b.append(zarry)
d29b = d29b.apply(pd.Series.interpolate)
d29b = d29b.reset_index(drop=True)
d29b = d29b[:383]

d30b = d30[:0]
for i in range(0, 40):
    d30b = d30b.append(d30.ix[i])
    d30b = d30b.append(zarry)
d30b = d30b.apply(pd.Series.interpolate)
d30b = d30b.reset_index(drop=True)
d30b = d30b[:383]
#d30b = abs(d30b)

d28b.head()
```

```python
# This code multiplies the total cross-section data by the quantum yield data and labels it appropriately.

dx28 = dx.copy()
dx28['sigma/cm2'] = dx['sigma/cm2']*d28b['phi']          
dx28.columns = ["lambda/A","[28]H2CO->HCO/H"]            # These lines need the [number] name of the photoreactions 
dx29 = dx.copy()
dx29['sigma/cm2'] = dx['sigma/cm2']*d29b['phi']
dx29.columns = ["lambda/A","[29]H2CO->CO/H2"]
dx30 = dx.copy()
dx30['sigma/cm2'] = dx['sigma/cm2']*d30b['phi']
dx30.columns = ["lambda/A","[30]H2CO->CO/H/H"]
dx28.head()
```

```python
# This code places the new data for each photoreaction into the old data.

df = pd.read_csv('/home/ag834/Data/Mainz Sigma/Sigma.dat', delimiter='   ') # This imports the original sigma table from ARGO
dxx28 = pd.DataFrame(np.zeros((10000, 1)), columns =  ["[28]H2CO->HCO/H"])             
dxx28['[28]H2CO->HCO/H'] = df['[28]H2CO->HCO/H']
dxx28 = dxx28[:2999]                               # This value is 1 less than the first lambda value
dxx28 = dxx28.append(dx28) 
dxx28 = dxx28.reset_index(drop=True)
dyy28 = pd.DataFrame(np.zeros((10000, 1)), columns =  ["[28]H2CO->HCO/H"])  # These lines also need the photoreaction name
dyy28['[28]H2CO->HCO/H'] = df['[28]H2CO->HCO/H']
dxx28 = dxx28.append(dyy28[3382:])                 # This value is the last lambda value
dxx28 = dxx28.drop('lambda/A',1)

dxx29 = pd.DataFrame(np.zeros((10000, 1)), columns =  ["[29]H2CO->CO/H2"])
dxx29['[29]H2CO->CO/H2'] = df['[29]H2CO->CO/H2']
dxx29 = dxx29[:2999]
dxx29 = dxx29.append(dx29) 
dxx29 = dxx29.reset_index(drop=True)
dyy29 = pd.DataFrame(np.zeros((10000, 1)), columns =  ["[29]H2CO->CO/H2"])
dyy29['[29]H2CO->CO/H2'] = df['[29]H2CO->CO/H2']
dxx29 = dxx29.append(dyy29[3382:])
dxx29 = dxx29.drop('lambda/A',1)
```

```python
# This code imports the new data into a dataframe of all the cross-section data.

dfnew = df.copy()
# H2CO Data is columns 28-33
Q = np.arange(28,142)         # This line needs to be (first updated photoreaction number:142)
W = np.arange(0,30)           # This line needs to be (0:last updated photreaction number)
dfnew.drop(dfnew.columns[Q], axis=1, inplace=True)
dfnew = dfnew.join(dxx28)
dfnew = dfnew.join(dxx29)
dfend = df.copy()
dfend.drop(dfnew.columns[W], axis=1, inplace=True)
dfnew = dfnew.join(dfend)
dfnew.head()
```
