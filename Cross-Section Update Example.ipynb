```python
# Import Mainz data for the cross sections of HCHO and plot to compare it to that found from ARGO.
# Redistribute data into 1 angstrom bins so it can be incorprated into ARGO.
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#%matplotlib inline
```

```python
### Original ARGO Sigma Data ###
df = pd.read_csv('/home/ag834/Data/Mainz Sigma/Sigma.dat', delimiter='   ')
OriginalSigma = df['[28]H2CO->HCO/H'] + df['[29]H2CO->CO/H2'] + df['[30]H2CO->CO/H/H'] + df['[31]H2CO->H2CO+'] + df['[32]H2CO->HCO+H'] + df['[33]H2CO->CO+H2']

plt.semilogy(df['lambda'], OriginalSigma)
plt.ylabel('$\sigma$ /$cm^2 $', fontsize =28)
plt.xlabel('Wavelength /$\AA$', fontsize =28)
plt.title('Plot of The Absorption Cross Section of HCHO according to ARGO as a function of Wavelength', fontsize =30, y=1.03)

ax = plt.gca()
ax.tick_params(width=2, labelsize =25)
ax.tick_params(length=5)

plt.xlim([0, 4000])                        
plt.ylim([1E-23, 1E-15]) 

plt.rcParams['figure.figsize'] = (15.0, 9.0)
plt.show()
```

```python
### Smith Data (300-340nm) ###
ds = pd.read_csv('/home/ag834/Data/Mainz Sigma/Smith 300-340nm 294K', delimiter='\t')
SmithSigma = abs(ds['sigma'])

plt.semilogy(df['lambda'], OriginalSigma, label='Smith Data', color='r', linewidth=2.5)
plt.semilogy(df['lambda'], OriginalSigma, label='Original Data', color='k', linewidth=2.5)
plt.ylabel('$\sigma$ /$cm^2 $', fontsize =28)
plt.xlabel('Wavelength /$\AA$', fontsize =28)
#plt.title('Plot of The Absorption Cross Section of HCHO as a function of Wavelength', fontsize =32, y=1.03)
plt.legend(loc='lower left', fontsize =28)
plt.semilogy(ds['lambda']*10, SmithSigma, label='Smith Data', color='r')
plt.semilogy(df['lambda'], OriginalSigma, label='Original Data', color='k', linewidth=2.5)

ax = plt.gca()
ax.tick_params(width=2, labelsize =25)
ax.tick_params(length=5)

plt.xlim([3000, 3400])                        
plt.ylim([1E-24, 1E-18])                    

plt.rcParams['figure.figsize'] = (15.0, 9.0)
plt.show()
```

```python
# In order to incorperate data, the bins need to be of width 1 angstrom (0.1nm).
# Method 1: Taking every 100th value
dsa = ds.iloc[::100, :]
dsa['lambda'] = (dsa['lambda'] + 0.004)
RSmithSigma1 = abs(dsa['sigma'])

plt.semilogy(df['lambda'], OriginalSigma, label='Binned Smith Data', color='r', linewidth=2)
plt.semilogy(df['lambda'], OriginalSigma, label='Original Data', color='k', linewidth=2)
plt.ylabel('$\sigma$ /$cm^2 $', fontsize =28)
plt.xlabel('Wavelength /$\AA$', fontsize =28)
#plt.title('Plot of The Absorption Cross Section of HCHO, taking every $100^{th}$ value', fontsize =31, y=1.03)
plt.legend(loc='lower left', fontsize =28)
plt.semilogy(df['lambda'], OriginalSigma, label='Original Data', color='k', linewidth=2)
plt.semilogy(dsa['lambda']*10, RSmithSigma1, label='Binned Smith Data', color='r')

ax = plt.gca()
ax.tick_params(width=2, labelsize =25)
ax.tick_params(length=5)

plt.xlim([3000, 3400])                        
plt.ylim([1E-24, 1E-18])                    

plt.rcParams['figure.figsize'] = (15.0, 9.0)
plt.show()
```

```python
# Method 2: Averaging over each bin of 100 values
for i in range(0, 383):
    dsa[i] = np.mean(ds['sigma'][i*100 : (i+1)*100])
    
dsab = dsa.iloc[4]
RSmithSigma2 = dsab[2:386]

plt.semilogy(df['lambda'], OriginalSigma, label='Binned Smith Data', color='r', linewidth=2)
plt.semilogy(df['lambda'], OriginalSigma, label='Original Data', color='k', linewidth=2)
plt.ylabel('$\sigma$ /$cm^2 $', fontsize =28)
plt.xlabel('Wavelength /$\AA$', fontsize =28)
#plt.title('Plot of The Absorption Cross Section of HCHO, averaging every $100$ values', fontsize =31, y=1.03)
plt.legend(loc='lower left', fontsize =28)
plt.semilogy(dsa['lambda'][0:383]*10, RSmithSigma2, label='Redistributed Smith Data', color='r')
plt.semilogy(df['lambda'], OriginalSigma, label='Original Data', color='k', linewidth=2)

ax = plt.gca()
ax.tick_params(width=2, labelsize =25)
ax.tick_params(length=5)

plt.xlim([3000, 3400])                        
plt.ylim([1E-24, 1E-18])                    

plt.rcParams['figure.figsize'] = (15.0, 9.0)
plt.show()
```

```python
# Data to be Added 3000 - 3382 A 
dx = pd.DataFrame(np.zeros((383, 2)), columns =  ["lambda/A", "sigma/cm2"])
dx['lambda/A'] = np.arange(3000,3383).reshape((383, 1))
dx['sigma/cm2'] = RSmithSigma2
dx.head()
```

```python
# Quantum yields from Mainz for reactions
# [28]H2CO->HCO/H   [29]H2CO->CO/H2   [30]H2CO->CO/H/H   [31]H2CO->H2CO+   [32]H2CO->HCO+H   [33]H2CO->CO+H2
d28 = pd.read_csv('/home/ag834/Data/Mainz Sigma/H + HCO Phi', delimiter='        ')
d29 = pd.read_csv('/home/ag834/Data/Mainz Sigma/H2 + CO Phi', delimiter='        ')
dt = pd.read_csv('/home/ag834/Data/Mainz Sigma/Total Phi', delimiter='        ')
d30 = dt.copy()
d30['phi'] = dt['phi'] - d28['phi'] - d29['phi']
d28 = d28[50:90]
d28 = d28.reset_index(drop=True)
d29 = d29[50:90]
d29 = d29.reset_index(drop=True)
d30 = d30[50:90]
d30 = d30.reset_index(drop=True)

zarry = pd.DataFrame(np.zeros((9, 2)), columns =  ["lambda", "phi"])
zarry[:] = np.nan

d28b = d28[:0]
for i in range(0, 40):
    d28b = d28b.append(d28.ix[i])
    d28b = d28b.append(zarry)
d28b = d28b.apply(pd.Series.interpolate)
d28b = d28b.reset_index(drop=True)
d28b = d28b[:383]

d29b = d29[:0]
for i in range(0, 40):
    d29b = d29b.append(d29.ix[i])
    d29b = d29b.append(zarry)
d29b = d29b.apply(pd.Series.interpolate)
d29b = d29b.reset_index(drop=True)
d29b = d29b[:383]

d30b = d30[:0]
for i in range(0, 40):
    d30b = d30b.append(d30.ix[i])
    d30b = d30b.append(zarry)
d30b = d30b.apply(pd.Series.interpolate)
d30b = d30b.reset_index(drop=True)
d30b = d30b[:383]
#d30b = abs(d30b)

d28b.head()
```

```python
dx28 = dx.copy()
dx28['sigma/cm2'] = dx['sigma/cm2']*d28b['phi']
dx28.columns = ["lambda/A","[28]H2CO->HCO/H"]
dx29 = dx.copy()
dx29['sigma/cm2'] = dx['sigma/cm2']*d29b['phi']
dx29.columns = ["lambda/A","[29]H2CO->CO/H2"]
dx30 = dx.copy()
dx30['sigma/cm2'] = dx['sigma/cm2']*d30b['phi']
dx30.columns = ["lambda/A","[30]H2CO->CO/H/H"]
dx28.head()
```

```python
# Plot to determine initial phis in model
OriginalSigma = df['[28]H2CO->HCO/H'] + df['[29]H2CO->CO/H2'] + df['[30]H2CO->CO/H/H'] + df['[31]H2CO->H2CO+'] + df['[32]H2CO->HCO+H'] + df['[33]H2CO->CO+H2']

plt.semilogy(df['lambda'], df['[28]H2CO->HCO/H'], label='$[28] H_2 CO \longrightarrow HCO/H$') 
plt.semilogy(df['lambda'], df['[29]H2CO->CO/H2'], label='$[29] H_2 CO \longrightarrow CO/H_2$') 
plt.semilogy(df['lambda'], df['[30]H2CO->CO/H/H'], label='$[30] H_2 CO \longrightarrow CO/H/H$') 
plt.semilogy(df['lambda'], df['[31]H2CO->H2CO+'], label='$[31] H_2 CO \longrightarrow H_ 2CO^+$') 
plt.semilogy(df['lambda'], df['[32]H2CO->HCO+H'], label='$[32] H_2 CO \longrightarrow HCO+H$') 
plt.semilogy(df['lambda'], df['[33]H2CO->CO+H2'], label='$[33] H_2 CO \longrightarrow CO+H_2$') 


plt.ylabel('$\sigma$ /$cm^2 $', fontsize =28)
plt.xlabel('Wavelength /$\AA$', fontsize =28)
plt.title('Plot of The Absorption Cross Sections multiplied by the Quantum Yields according to ARGO', fontsize =30, y=1.03)
plt.legend(loc='upper right', fontsize =28)

ax = plt.gca()
ax.tick_params(width=2, labelsize =25)
ax.tick_params(length=5)

plt.xlim([0, 4000])                        
plt.ylim([1E-23, 1E-15])                    

plt.rcParams['figure.figsize'] = (15.0, 9.0)
plt.show()
```

```python
# Create new Dataframe with old data surrounding the new RSmithSigma2
dxx28 = pd.DataFrame(np.zeros((10000, 1)), columns =  ["[28]H2CO->HCO/H"])
dxx28['[28]H2CO->HCO/H'] = df['[28]H2CO->HCO/H']
dxx28 = dxx28[:2999]
dxx28 = dxx28.append(dx28) 
dxx28 = dxx28.reset_index(drop=True)
dyy28 = pd.DataFrame(np.zeros((10000, 1)), columns =  ["[28]H2CO->HCO/H"])
dyy28['[28]H2CO->HCO/H'] = df['[28]H2CO->HCO/H']
dxx28 = dxx28.append(dyy28[3382:])
dxx28 = dxx28.drop('lambda/A',1)

dxx29 = pd.DataFrame(np.zeros((10000, 1)), columns =  ["[29]H2CO->CO/H2"])
dxx29['[29]H2CO->CO/H2'] = df['[29]H2CO->CO/H2']
dxx29 = dxx29[:2999]
dxx29 = dxx29.append(dx29) 
dxx29 = dxx29.reset_index(drop=True)
dyy29 = pd.DataFrame(np.zeros((10000, 1)), columns =  ["[29]H2CO->CO/H2"])
dyy29['[29]H2CO->CO/H2'] = df['[29]H2CO->CO/H2']
dxx29 = dxx29.append(dyy29[3382:])
dxx29 = dxx29.drop('lambda/A',1)
```

```python
# Plots to compare Sigma*Phi for reactions 28 and 29
plt.semilogy(df['lambda'], df['[28]H2CO->HCO/H'], label='$[28] H_2 CO \longrightarrow HCO/H$ Original') 
plt.semilogy(df['lambda'], dxx28['[28]H2CO->HCO/H'], label='$[28] H_2 CO \longrightarrow HCO/H$ New') 

plt.ylabel('$\sigma$ /$cm^2 $', fontsize =28)
plt.xlabel('Wavelength /$\AA$', fontsize =28)
plt.title('Plot of The Original and New Absorption Cross Sections multiplied by the Quantum Yields', fontsize =30, y=1.03)
plt.legend(loc='lower left', fontsize =28)

ax = plt.gca()
ax.tick_params(width=2, labelsize =25)
ax.tick_params(length=5)

plt.xlim([3000, 3382])                        
plt.ylim([5E-24, 1E-19])                    

plt.rcParams['figure.figsize'] = (15.0, 9.0)
plt.show()
```
